```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('jersey-rainfall-1984-to-2021.csv',index_col=0)
df.head()
```

```python
df.plot(y="YearTotalmm")
plt.show()
```

```python
df.plot(y="YearTotalmm",kind='area')
plt.show()
```

```python
df.plot(y="Jan")
plt.show()
```

```python
df.plot(y=['Dec',"Jan",'Feb'],kind='area')
plt.show()
```

```python
df.plot()
plt.show()
```

```python
df.drop(columns=['YearTotalmm']).plot(kind='area')
plt.show()
```

```python
df.T.head()
```

```python
df.drop(columns=['YearTotalmm']).T.plot(y=[1901,1902,1903,2019,2020],kind='bar')
plt.show()
```
